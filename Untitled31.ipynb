```python
def initialize_chatbot(path):
    path='/content/urlbooks (1) (1).csv'
    df = pd.read_csv(path)
    links = df['Link'].tolist()
    loaders = UnstructuredURLLoader(urls=links)
    data = loaders.load()
    text_splitter = CharacterTextSplitter(separator='\n', chunk_size=1000, chunk_overlap=200)
    docs = text_splitter.split_documents(data)
    embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
    os.environ["OPENAI_API_KEY"]="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"
    faiss = 'faiss_db'
    vectorStore_openAI = FAISS.from_documents(docs, embeddings)
    vectorStore_openAI.save_local(faiss)
    reload_faiss = FAISS.load_local(faiss, embeddings)
    llm = ChatOpenAI(temperature=0, model_name="gpt-3.5-turbo")
    chain = RetrievalQAWithSourcesChain.from_llm(llm=llm, retriever=reload_faiss.as_retriever())
    return chain
```

```python
def conversation_flow(history, user_input, chain):
    # Generate conversational openers and closers
    if not history:
        welcome_message = "Hi there! I'm your friendly book expert, ready to answer any questions you have about books in our collection. What can I help you with today?"
        return update_history(history, {"role": "system", "content": welcome_message})
    elif user_input.lower() == "bye" or user_input.lower() == "thanks":
        farewell_message = "It was a pleasure chatting with you about books! Come back anytime you have more questions. Happy reading!"
        return update_history(history, {"role": "system", "content": farewell_message}), True
    # Process user input and generate response, considering history
    messages = update_history(history, {"role": "user", "content": user_input})
    bot_response = get_chatgpt_response(messages, chain)
    return update_history(messages, {"role": "system", "content": bot_response}), False



```

```python
# Function for updating conversation history
def update_history(history, message):
    if not history:
        history = []
    history.append(message)
    return history

# Function for generating chatbot response using the chain
def get_chatgpt_response(messages, chain):
    response = chain({"question": messages}, return_only_outputs=True)
    return response.choices[0].message.content

# Example usage:
path = "/content/urlbooks (1) (1).csv"  # Replace with your CSV path
chain = initialize_chatbot(path)
history = None  # Initialize conversation history

while True:
    user_input = input("You: ")
    conversation_ended, history = conversation_flow(history, user_input, chain)
    if conversation_ended:
        break
    print("Chatbot:", history[-1]["content"])
```
